
# Task 1: Predicting the scores of students based on the number of hours of study by using linear regression.

```python
# importing important libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Reading the file
data="http://bit.ly/w-data"
new_data=pd.read_csv(data)
print("Data imported successfully")
new_data.head(10)
```

```python
# Plotting the dataset for manually finding the ralationship between the data.
new_data.plot(x="Hours", y="Scores", style='o')
plt.title("Hours Vs Percentage")
plt.xlabel("Hours Studied")
plt.ylabel("Percentage Scores")
plt.show()
```

```python
#Dividing the data into "attributes" (inputs) and "labels" (outputs).
X=new_data.iloc[:, :-1].values
Y=new_data.iloc[:, 1].values
```

```python
# Splitting the data into training and test sets by using Scikit-Learn's built-in train_test_split() method:
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X,Y, test_size=0.2, random_state=0)
```

```python
# Training the algorithm
from sklearn.linear_model import LinearRegression
regg=LinearRegression()
regg.fit(X_train,Y_train)
print("Training Complete")
```

```python
# Plotting the regression line and test data
line=regg.coef_*X+regg.intercept_
plt.scatter(X,Y)
plt.plot(X,line)
plt.show()
```

```python
# Predicting the scores
print(X_test)
Y_pred=regg.predict(X_test)
print(Y_pred)
```

```python
# comparing actual Vs Predicted result
df=pd.DataFrame({'Actual': Y_test ,'Predicted': Y_pred})
df
```

```python
# Testing with own data
hours=float(input("Enter the hours:"))
own_pred=regg.predict([[hours]])
print("number of hours={}".format(hours))
print("Predicted score={}".format(own_pred[0]))
```

```python
#evaluating the model by using mean absolute error
from sklearn import metrics
print("Mean Absolute error:", metrics.mean_absolute_error(Y_test,Y_pred))
```
